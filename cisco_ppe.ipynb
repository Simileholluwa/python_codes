```python
secret_number = 777

print(
"""
+===============================+
| Welcome to my game, muggle!   |
| Enter an Integer number       |
| and guess what number I've    |
| picked for you.               |
| So, what is the number        |
+===============================+
"""
)

number = input("Enter the number you think it is: ")

if number.isdigit():
    while number != 777:  
        if number != secret_number:
            print("Ha ha! You are stuck in my loop!")
        number = int(input("Enter the number you think it is: "))

    print("Well done, muggle! You are free now")

else: 
    print("You can only enter integer values, muggle! Try again.")

```

```python
try:
	c = int(input("Enter a non negative and non zero number: "))
	
	steps = 0
	
	if c < 0:
		print("Kindly enter a non negative and non zero number ")
	
	else:
		while c != 1:
			if c % 2 == 0:
				steps += 1
				c /= 2
			elif c % 2 != 0:
				steps += 1
				c = 3 * c + 1
			print("Current value: ", c)
			print("\n")
		else :
			print("Total number of steps: ", steps)
			
except:
	print("Ensure your inputs are integer values only")
```

```python
my_list = []
swapped = True
num = int(input("How many elements do you want to sort? "))

for _ in range(num):
    val = float(input("Enter the list elements: "))
    my_list.append(val)

while swapped:
    swapped = False
    for i in range(len(my_list) - 1):
        if my_list[i] > my_list[i + 1]:
            swapped = True
            my_list[i], my_list[i + 1] = my_list[i + 1], my_list[i]

print("Sorted list:", my_list)
```

```python
number_of_elem = int(input("How many elements are in the list? "))

my_list = []

new_list = []

for _ in range(number_of_elem):
    elements = float(input("Enter the elements of the list"))
    my_list.append(elements)

for elem in my_list:
    if elem not in new_list:
        new_list.append(elem)

print("The list you entered:" ,my_list)
print("The list without repeated numbers:" ,new_list)
```

```python
def is_prime(num):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                print(num, "is not a prime number")
                break
        else:
            print(num, "is a prime number")

    else:
        print("Your input must be greater than 1")


for i in range(1, 20):
    if is_prime(i + 1):
        print(i + 1, end = "")
print()
```
